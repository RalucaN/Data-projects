```python
import json
import pandas as pd
from datetime import datetime

with open('dict.json', 'r') as myfile:
    data=myfile.read()

steam_data = json.loads(data)
```

<h3>Checking what type of data is <i>steam_data</i></h3>

```python
print(type(steam_data))
```

<h3>Exploring the dictionary <i>steam_data</i> by printing they keys and identifying way to parse it</h3>

```python
for key in steam_data:
    print(key)
```

<h4>a.)Removing <i>query_summary</i> and <i>cursors</i> keys</h4>

```python
steam_data = {key: steam_data[key] for key in steam_data if key not in ["query_summary","cursors"]}
for key in steam_data:
    print(key)
```

```python
def parse(dic):
    for k, v in dic.items():
        if isinstance(v, dict):
            for p in parse(v):
                yield [k] + p
        else:
            yield [k, v]

reviews_list = list(parse(steam_data))
```

```python
reviews_list
```

```python
type(reviews_list)
```

<h3>Removing the <i>author</i> and <i>reviews</i> value as it is not necessary in the analysis</h3>

```python
for i in reviews_list:
  try:
    i.remove("author")
  except ValueError:
    pass
```

```python
reviews_list
```

```python
for i in reviews_list:
  try:
    i.remove("reviews")
  except ValueError:
    pass

reviews_list
```

<h3>Transforming <i>reviews_list</i> into a dataframe</h3>

```python
data1=pd.DataFrame(reviews_list, columns=["id", "column_name", "value"])
reviews_table=data1.pivot(index="id", columns="column_name", values="value")
reviews_table
```

<h3>Cleaning the dataframe: transforming timestamps to datetime, indicating the type of data in each column</h3>

```python
reviews_table.info(verbose=True)
```

```python
reviews_table["comment_count"]=reviews_table["comment_count"].astype(int) #number of comments posted on this review
reviews_table["num_games_owned"]=reviews_table["num_games_owned"].astype(int) #number of games owned by the user
reviews_table["num_reviews"]=reviews_table["num_reviews"].astype(int) #number of reviews written by the user
reviews_table["received_for_free"]=reviews_table["received_for_free"].astype(bool) #true if the user checked a box saying they got the app for free
reviews_table["steam_purchase"]=reviews_table["steam_purchase"].astype(bool) #true if the user purchased the game on Steam
reviews_table["voted_up"]=reviews_table["voted_up"].astype(bool) #true means it was a positive recommendation
#reviews_table["votes_funny"]=reviews_table["votes_funny"].astype(int) #the number of users that found this review funny
reviews_table["votes_up"]=reviews_table["votes_up"].astype(int) #the number of users that found this review helpful
reviews_table["weighted_vote_score"]=reviews_table["weighted_vote_score"].astype(float) # helpfulness score
reviews_table["written_during_early_access"]=reviews_table["written_during_early_access"].astype(bool) #true if the user posted this review while the game was in Early Access
```

```python
reviews_table.info(verbose=True)
```

```python
reviews_table["last_played"]=pd.to_datetime(reviews_table["last_played"], unit='s')
reviews_table["last_played"]
```

```python
reviews_table["date_created"]=pd.to_datetime(reviews_table["timestamp_created"], unit='s')
reviews_table["date_created"]
```

```python
reviews_table["date_updated"]=pd.to_datetime(reviews_table["timestamp_updated"], unit='s')
reviews_table["date_updated"]
```

```python
reviews_table.info(verbose=True)
```

```python
reviews_table["hours_played"]=reviews_table["playtime_forever"].astype(int)/60
reviews_table["hours_played"] #number of hours played by used
```

```python
reviews_table.to_csv("reviews_table.csv", encoding='utf-8')
```
